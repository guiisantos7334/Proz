```python
def calculadora(parametro1, parametro2, parametro3):
  if(parametro3==1):
    return (parametro1+parametro2)
  elif(parametro3==2):
    return (parametro1-parametro2)
  elif(parametro3==3):
    return (parametro1*parametro2)
  elif(parametro3==4):
    return (parametro1/parametro2)

numero3=1
while(numero3!=0):
  print("Selecione a operação desejada: \n1: Soma \n2: Subtração \n3: Multiplicação \n4: Divisão \n0: Sair")
  numero3=int(input())
  if(numero3==0):
    print("Você saiu")
  elif(numero3>4):
    print("Essa opção não existe\n")
  else:
    print("Selecione o primero numero\n")
    numero1=int(input())
    print("Selecione o segundo numero\n")
    numero2=int(input())
    resultado=(calculadora(numero1,numero2,numero3))
    print("Resultado:", resultado)


```
